```python
import plotnine

from plotnine import ggplot, aes, geom_line, geom_point, geom_bar, stat_bin, geom_histogram
from plotnine import geom_boxplot, scale_x_timedelta, labs, facet_grid, theme_dark
```

```python
from plotnine.data import economics

economics.head()
```

```python
ggplot(economics) + aes(x='date', y='pop') + geom_line()
```

```python
from plotnine.data import mpg

mpg.head()
```

```python
ggplot(mpg) + aes(x='class', y='hwy') + geom_point()
```

```python
ggplot(mpg) + aes(x="class") + geom_bar()
```

```python
from plotnine.data import huron

huron.head()
```

```python
ggplot(huron) + aes(x="level") + stat_bin(bins=10) + geom_bar()
```

```python
ggplot(huron) + aes(x="level") + geom_histogram(bins=10)
```

```python
ggplot(huron) + aes(x='factor(decade)', y='level') + geom_boxplot()
```

```python
(
    ggplot(economics)
    + aes(x="date", y="pop")
    + scale_x_timedelta(name="Years since 1970")
    + labs(title="Population Evolution", y="Population")
    + geom_line()
)
```

```python
(
    ggplot(mpg)
    + facet_grid(facets="year~class")
    + aes(x="displ", y="hwy")
    + labs(
        x="Engine Size",
        y="Miles per Gallon",
        title="Miles per Gallon for Each Year and Vehicle Class",
    )
    + geom_point()
)
```

```python
(
    ggplot(mpg)
    + facet_grid(facets="year~class")
    + aes(x="displ", y="hwy")
    + labs(
        x="Engine Size",
        y="Miles per Gallon",
        title="Miles per Gallon for Each Year and Vehicle Class",
    )
    + geom_point()
    + theme_dark()
)
```

```python
(
    ggplot(mpg)
    + aes(x="cyl", y="hwy", color="class")
    + labs(
        x="Engine Cylinders",
        y="Miles per Gallon",
        color="Vehicle Class",
        title="Miles per Gallon for Engine Cylinders and Vehicle Classes",
    )
    + geom_point()
)
```
